```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# print how many rows are there in the table named 'master'
QUERY = "SELECT COUNT(*)FROM master"

# the read_sql_query runs the sql query and returns the data as a dataframe
sql_df = pd.read_sql_query(QUERY,conn)
sql_df.head()
```

```python
# print all the tables names in the database
QUERY = "SELECT name as Table_Name FROM sqlite_master WHERE type = 'table'"
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

```python
table_name = 'master'  # the table you wish to describe

QUERY = "SELECT sql FROM sqlite_master WHERE name= '{}'".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

#Pie Chart

```python
Query = "select * from sqlite_master"
pd.read_sql_query(Query,conn)
```

```python
Query= "select * from DatabaseDesireNextYear"
sql_df = pd.read_sql_query(Query,conn)
sql_df.head()
```

```python
Query= "select DatabaseDesireNextYear from DatabaseDesireNextYear group by DatabaseDesireNextYear order by Respondent"
pd.read_sql_query(Query,conn)
```

```python
#Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.
#df_db = sql_df.groupby(['DatabaseDesireNextYear'], axis =0)
df_db = sql_df.groupby('DatabaseDesireNextYear', axis=0).count()
df_db.sort_values(["Respondent"], ascending=False, inplace=True)
#df_db.sort(['Respondent'], ascendinng = False, inplace= True)
df_top5db = df_db.head()
df_top5db
```

```python
df_top5db['Respondent'].plot(kind='pie', autopct='%1.1f%%')
plt.show()
```

```python
query = "select * from LanguageDesireNextYear"
df_langdb = pd.read_sql_query(query,conn)
df_langdb.head()
```

```python
lang_group = df_langdb.groupby(df_langdb['LanguageDesireNextYear'], axis = 0).count()
lang_group.sort_values(['Respondent'], ascending =False, inplace= True)
lang_group.head()
```

#Stacked Chart

```python
#Create a stacked chart of median WorkWeekHrs and CodeRevHrs for the age group 30 to 35.
new_df = df.loc[(df['Age'] >= 30) & (df['Age'] <= 35)]
new_df = new_df[['WorkWeekHrs','CodeRevHrs','Age']]
new_df.head()
```

```python
new_df[['WorkWeekHrs','CodeRevHrs']].median()
```

```python
x = new_df['Age']
y1 = new_df['WorkWeekHrs'].median()
y2 = new_df['CodeRevHrs']
plt.bar(x, y1, color='g')
plt.bar(x, y2, bottom=y1, color='y')
plt.show()
```
